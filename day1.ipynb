# Factorial program using Recurtion

```python
def fact(n):
    if n == 0:
        return 1
    else:
        return n*fact(n-1)

n = int(input('Enter the number: '))
res = fact(n)
print(res)
```

# Threading in python

```python
# Process:- A process is an executable instance of a computer program
# Threading:- A thread is a sequeance of instruction in program that can be executed independentely of the remaining program.
```

```python
from threading import *
class MyThread(Thread):
    def run(self):
        for i in range(5):
            print("\n this is child thread")
t=MyThread()
t.start()
for i in range(5):
    print("\n this is the main thread")
```

```python
from threading import*
class Demo:
    def show(self):
        for i in range(5):
            print("\n this is a child thread")
obj=Demo()
t=Thread(target=obj.show())
t.start()
for i in range(5):
    print("\n this is parent thread")
    
```

# Multithreading

```python
# Multithreading is a model where multiple threads within a process execute independently while sharing the same resources.
```

```python
from threading import *
import time
class Demo:
    def num(self):
        for i in range(1,6):
            print("the number is",i)
            time.sleep(1)
    def double(self):
        for i in range(1,6):
            print("the double of number is",2*i)
            time.sleep(1)
    def square(self):
        for i in range(1,6):
            print("the square number is",i*i)
            time.sleep(1)
obj=Demo()
t1=Thread(target=obj.num)
t2=Thread(target=obj.double)
t3=Thread(target=obj.square)

t1.start()
time.sleep(0.2)
t2.start()
time.sleep(0.2)
t3.start()

t1.join()
t2.join()
t3.join()

print("this is main thread")


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for contour map
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10), dpi=150)
axes = axes.flatten()  # Flatten the axes array for easier indexing

# Line Plot
line_y = np.sin(x)
axes[0].plot(x, line_y, color='black')
axes[0].set_title('Line Plot')
axes[0].grid(True)

# Scatter Plot
scatter_y = np.random.randn(100)
axes[1].scatter(x[:100], scatter_y, color='green')
axes[1].set_title('Scatter Plot')
axes[1].grid(True)

# Bar Plot
bar_categories = ['A', 'B', 'C']
bar_values = [10, 20, 15]
axes[2].bar(bar_categories, bar_values, color='blue')
axes[2].set_title('Bar Plot')
axes[2].grid(True)

# Histogram
hist_data = np.random.randn(1000)
axes[3].hist(hist_data, bins=30, color='purple')
axes[3].set_title('Histogram')
axes[3].grid(True)

# Box Plot
box_data = [np.random.randn(100), np.random.randn(100) + 1, np.random.randn(100) + 2]
axes[4].boxplot(box_data)
axes[4].set_title('Box Plot')
axes[4].grid(True)

# Contour Map
contour = axes[5].contourf(X, Y, Z, cmap='viridis')
fig.colorbar(contour, ax=axes[5])
axes[5].set_title('Contour Map')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the website to scrape
url = 'https://www.bbc.com/news'

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all article tags (assuming articles are in <h3> tags with a specific class)
    articles = soup.find_all('h3', class_='gs-c-promo-heading__title')

    # Iterate through the articles and extract titles and links
    for article in articles:
        title = article.get_text()
        link = article.find_parent('a')['href']
        if not link.startswith('http'):
            link = 'https://www.bbc.com' + link  # Complete relative URLs

        print(f'Title: {title}')
        print(f'Link: {link}')
        print()
else:
    print(f'Failed to retrieve the webpage. Status code: {response.status_code}')
```

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the website to scrape
url = 'http://books.toscrape.com/'

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all book items (each book is in an article tag with class 'product_pod')
    books = soup.find_all('article', class_='product_pod')

    # Iterate through the books and extract titles, prices, and availability
    for book in books:
        # Extract the title
        title = book.h3.a['title']
        
        # Extract the price
        price = book.find('p', class_='price_color').text
        
        # Extract the availability
        availability = book.find('p', class_='instock availability').text.strip()
        
        print(f'Title: {title}')
        print(f'Price: {price}')
        print(f'Availability: {availability}')
        print()
else:
    print(f'Failed to retrieve the webpage. Status code: {response.status_code}')
```
